# 1.-Concepto de Data Science
La selecci칩n y filtrado condicional en Pandas utilizan m치scaras booleanas para seleccionar subconjuntos de datos basados en condiciones l칩gicas. Son cruciales en Data Wrangling porque permiten aislar registros espec칤ficos, detectar errores o anomal칤as (como valores fuera de rango) y enfocar el an치lisis en datos relevantes, mejorando la eficiencia en la preparaci칩n de datasets grandes.

## Sitanxis

M치scara booleana simple (condici칩n sobre una columna)

`mascara = df['columna'] > valor`

 Filtrado: aplicar la m치scara al DataFrame

`df_filtrado = df[mascara]`

 Condiciones m칰ltiples: usar & (AND), | (OR), ~ (NOT)

`df_filtrado = df[(df['col1'] > val1) & (df['col2'] == val2)]`

### Ejemplo Practico

Usaremos un DataFrame de ventas con variaciones en precios y unidades para filtrar productos con ventas altas (unidades > 150) o precios bajos (< 60), resolviendo el problema de aislar datos de inter칠s para un an치lisis focalizado, como identificar best-sellers o descuentos.








````python
import pandas as pd

# Crear DataFrame de ejemplo
data_ventas = {
    'Producto': ['A', 'B', 'C', 'D'],
    'Unidades': [100, 200, 150, 180],
    'Precio': [50.0, 75.0, 55.0, 65.0]
}
df_ventas = pd.DataFrame(data_ventas)

# Filtrado simple: unidades > 150
mascara_alta_venta = df_ventas['Unidades'] > 150
df_alta_venta = df_ventas[mascara_alta_venta]

# Filtrado m칰ltiple: unidades > 150 O precio < 60
df_interes = df_ventas[(df_ventas['Unidades'] > 150) | (df_ventas['Precio'] < 60)]

# Mostrar resultados
print("Filtrado simple (unidades > 150):")
print(df_alta_venta)

print("\nFiltrado m칰ltiple (unidades > 150 O precio < 60):")
print(df_interes)
````

# 2.-Concepto Data Science(dropna)(fillna)
El manejo de valores nulos en Pandas con dropna() y fillna() es esencial para limpiar datasets incompletos. Son cruciales en Data Wrangling porque los nulos pueden distorsionar an치lisis; dropna() elimina filas/columnas con nulos para datos limpios, mientras fillna() los reemplaza (con medias, ceros, etc.) para preservar m치s informaci칩n y evitar sesgos en modelos o res칰menes.

## Sintaxis
Eliminar filas con al menos un nulo (axis=0 por default)

`df_limpio = df.dropna()`

 Eliminar columnas con nulos

`df_limpio = df.dropna(axis=1)`

Rellenar nulos con un valor espec칤fico (e.g., 0)

`df_rellenado = df.fillna(0)`

Rellenar con media de la columna

`df_rellenado = df.fillna(df.mean(numeric_only=True))`

### Ejemplo Pra치ctico con Dataframe

Usaremos un DataFrame de ventas con nulos en 'Unidades' y 'Precio' para simular datos reales incompletos. Aplicamos dropna() para eliminar filas inv치lidas y fillna() para rellenar con medias, resolviendo el problema de preparaci칩n para an치lisis sin perder todo el dataset.

#### Output Esperado
(medias: Unidades=(100+150+200)/3=150, Precio=(50+75+60)/3=61.666...):

````python
# Crear DataFrame con nulos
data_ventas = {
    'Producto': ['A', 'B', 'C', 'D'],
    'Unidades': [100, None, 150, 200],
    'Precio': [50.0, 75.0, None, 60.0]
}
df_ventas = pd.DataFrame(data_ventas)

# Eliminar filas con nulos
df_sin_nulos = df_ventas.dropna()

# Rellenar nulos con la media de cada columna num칠rica
df_rellenado = df_ventas.fillna(df_ventas.mean(numeric_only=True))

# Mostrar resultados
print("Original con nulos:")
print(df_ventas)

print("\nSin nulos (dropna):")
print(df_sin_nulos)

print("\nRellenado con medias (fillna):")
print(df_rellenado)
````

# 3.- Concepto DataScience (Columnas)
La creaci칩n de nuevas columnas en Pandas mediante vectorizaci칩n permite calcular m칠tricas derivadas de columnas existentes de forma eficiente sin bucles. Es crucial en Data Wrangling porque transforma datos crudos en features m치s 칰tiles para an치lisis, como ratios o categor칤as, optimizando el rendimiento en datasets grandes al usar operaciones vectorizadas de NumPy subyacentes.

## Sitaxis
 Nueva columna simple: operaci칩n aritm칠tica

`df['nueva_col'] = df['col1'] + df['col2']`

 Usando funciones: apply() para l칩gica compleja (pero preferir vectorizaci칩n)

`df['nueva_col'] = df['col1'].apply(lambda x: x * 2 if x > 0 else 0)`

 Condicionales vectorizadas: np.where() o operaciones booleanas
import numpy as np

`df['nueva_col'] = np.where(df['col1'] > valor, 'Alto', 'Bajo')`

### Ejemplo
Usaremos un DataFrame de ventas para crear una nueva columna 'Ingreso_Total' (Unidades * Precio) y otra 'Categoria_Precio' basada en umbrales, resolviendo el problema de generar m칠tricas derivadas para an치lisis como rentabilidad o segmentaci칩n.
python

````python
#Datos de ventas
import pandas as pd
import numpy as np
# Crear DataFrame de ejemplo
data_ventas = {
    'Producto': ['A', 'B', 'C'],
    'Unidades': [100, 200, 150],
    'Precio': [50.0, 75.0, 60.0]
}
df_ventas1=pd.DataFrame(data_ventas)
#Nuvea columna vectorizada: Ingreso_total=Unidades*precio
df_ventas1['Ingreso_Total']=df_ventas1['Unidades']*df_ventas1['Precio']
#Nueva columna vectorizada: Categoria_Precio basada en Precio>60
df_ventas1['Categoria_Precio']=np.where(df_ventas1['Precio']>60,'Alto','Bajo')
print(df_ventas1)
````

# 4.- Concepto DataScience
La agrupaci칩n y agregaci칩n en Pandas con groupby() permiten resumir datos por categor칤as, calculando m칠tricas como sumas, medias o conteos. Son clave en Data Wrangling porque transforman datos granulares en insights agregados, facilitando an치lisis como promedios por grupo o detecci칩n de patrones en subconjuntos, esenciales para reportes o modelado.

##Sintaxis
Agrupar por una columna y agregar (e.g., media)

`df_agrupado = df.groupby('columna_categoria')['columna_numerica'].mean()`

Agregaciones m칰ltiples: usar agg() con diccionario

```
df_agrupado = df.groupby('columna_categoria').agg({
    'col_numerica1': 'sum',
    'col_numerica2': ['mean', 'count']
})
```
Resetear 칤ndice para DataFrame plano

`df_agrupado = df_agrupado.reset_index()`

#### Ejercicio Practico
Usaremos un DataFrame de ventas por regi칩n para agrupar por 'Region' y calcular total de unidades vendidas y precio promedio, resolviendo el problema de resumir datos dispersos para an치lisis regional como identificaci칩n de mercados fuertes.

````python
import pandas as pd
data_ventas = {
    'Producto': ['A', 'B', 'A', 'C', 'B'],
    'Region': ['Norte', 'Sur', 'Norte', 'Este', 'Sur'],
    'Unidades': [100, 200, 150, 120, 180],
    'Precio': [50, 75, 55, 60, 70]
}
df_ventas = pd.DataFrame(data_ventas)

#Agrupar por "Region" y agregar: suma de unidades,media de precio
df_resumen=df_ventas.groupby('Region').agg({
    'Unidades':'sum',
    'Precio':'mean'
}).reset_index()
print("Df de datos\n",df_ventas)
print("Df de goupby\n",df_resumen)
````
